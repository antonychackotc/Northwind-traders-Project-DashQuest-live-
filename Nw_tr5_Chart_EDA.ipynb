```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create cross-tab with hierarchical row index (country and companyName)
cross_tab = pd.crosstab(
    [df['country'], df['companyName']],
    df['categoryName']
)

# 2. Display cross-tabulation
print(cross_tab)

# 3. Visualize with a heatmap
plt.figure(figsize=(14, 8))  # Adjust figure size as needed
sns.heatmap(cross_tab, annot=True, cmap="YlGnBu", fmt="d", linewidths=.5)
plt.title("Cross-Tabulation: Country & Company vs Category Name")
plt.xlabel("Category Name")
plt.ylabel("Country & Company")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.yticks(rotation=0)  # Keep y-axis labels readable
plt.tight_layout()
plt.show()
```

**United Packages from USA are highly Contributed for northerwind Traders especially ( Beverages and Seafood )**

**Speedy Express From USA 2nd Most Higly Contributed for Northwind Traders**
