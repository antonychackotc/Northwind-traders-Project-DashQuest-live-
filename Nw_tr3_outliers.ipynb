```python
import pandas as pd

def display_outliers_summary(outliers_summary):
    # Prepare a DataFrame for better visualization
    rows = []
    for col, info in outliers_summary.items():
        rows.append([
            col,
            info["Lower Bound Outliers"],
            info["Upper Bound Outliers"],
            info["Lower Outlier Values"],
            info["Upper Outlier Values"]
        ])

    outliers_df = pd.DataFrame(
        rows,
        columns=[
            "Column",
            "Lower Outlier Count",
            "Upper Outlier Count",
            "Lower Outlier Values",
            "Upper Outlier Values"
        ]
    )
    return outliers_df

# Example usage
outliers_summary = detect_outliers(df)
outliers_table = display_outliers_summary(outliers_summary)
```
