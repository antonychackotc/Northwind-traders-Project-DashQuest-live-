# **(ii) Merge**

# **1st merge (df1 and df2)**

```python
df8=pd.merge(df1,df2, on="orderID")
df8.head()
```

```python
df8.shape
```

# **2nd Merge (df3 and df8 [df1 & df2])**

```python
df9=pd.merge(df3,df8, on="customerID")
df9.head()
```

```python
df9.shape
```

# **3rd Merge (df4 and df9 - [df1,df2,df3])**

```python
df10=pd.merge(df4,df9, on="productID")
df10.head()
```

```python
df10.shape
```

# **4th Merge ( df5 and df10 - [ df1,df2,df3,df4 ])**

```python
df11=pd.merge(df5,df10, on="categoryID")
df11.head()
```

```python
df11.shape
```

# **5th Merge - ( df6 and df11 [ df1,df2,df3,df4,df5 ] )**

```python
df12=pd.merge(df6,df11, on="employeeID")
df12.head()
```

```python
df12.shape
```

# **6th Merge ( df7 and df12 [ df1,df2,df3,df4,df5,df6 ]**

```python
df13=pd.merge(df7,df12, on="shipperID")
df13.head()
```

```python
df13.shape
```

```python
df14 = df13.copy()
```

```python
df14.head()
```

```python
df14.columns
```

```python
# Merge or coalesce columns, taking '_x' where present, else '_y'
for col in set(col.rstrip('_x') for col in df14.columns if col.endswith('_x')):
    df14[col] = df14[f"{col}_x"].combine_first(df14[f"{col}_y"])
    df14.drop(columns=[f"{col}_x", f"{col}_y"], inplace=True, errors='ignore')
```

```python
df14.head()
```

```python
df14.columns
```

```python
df14.to_csv('northwind_traders')
```
